### Numpy Functions

```python
import numpy as np
```

```python
#Sort
a=np.array([1,8,3,6,2,9,5])

print(np.sort(a))
print(np.sort(a,kind="quicksort"))
print(np.sort(a,kind="mergesort")) #Kind provide us the option to chooese different types of sort
```

```python
a=np.random.randint(1,100,12).reshape(4,3)
print(a)
print("\n")
print(np.sort(a,axis=0))  #Column wise sort
print("\n")
print(np.sort(a,axis=1))  #Row wise sort
```

```python
#Append
a=np.array([1,8,3,6,2,9,5])
b=np.append(a,50)   #Append does not change the original array.
print(b)
```

```python
b=np.array([[12, 52, 42, 6],
        [29, 18, 47, 55],
        [61, 93, 83, 9],
        [38, 63, 44, 85],
        [ 8, 87, 31, 72],
        [40, 71, 2, 7]])

c=np.append(b,np.zeros(shape=(np.shape(b)[0],1)),axis=1)
print(c)
```

```python
#Concatenate 
c=np.arange(1,7).reshape(2,3)
d=np.arange(7,13).reshape(2,3)
print(np.concatenate((c,d),axis=1))
print("\n")
print(np.concatenate((c,d),axis=0))
```

```python
#Unique

a=[1,5,2,3,1,4,5,3,6,6]
np.unique(a)   #Returns the unique value
```

```python
#expand_dims

a=np.arange(1,13)
print(np.shape(a))

b=np.expand_dims(a,axis=0)  
print(np.shape(b))

c=np.expand_dims(a,axis=1)
print(np.shape(c))

#Important - expand_dims does NOT add rows or columns â€” it inserts a 1 at the given index in the shape (tuple).
```

```python
a=np.random.randint(1,100,20)
print(a)

#If I want all the numbers greater than 50

print(np.where(a>50))   #It provides the index positions. Not the values
```

```python
#If I want to replace all the numbers greater than 50
a=np.random.randint(1,100,20)
print(a)

print(np.where(a>50,0,a))  #It require 3 things, (condition, if true then, if false then)
#In our code we are saying if a>50, the n put 0 if not then the same no.
```

```python
#cumsum

a=np.arange(1,21)
print(np.cumsum(a)) #The cumsum returns the cumulative 
```

```python
a=np.arange(1,21)
print(np.cumprod(a))  #It returns the cumulative product
```

```python
#percentile

a=np.random.randint(1,100,20)
print(a)
np.percentile(a,98)  
#Percentile can be explained using the example- Let's say someone got 96 percentile in their result
#it means that person is ahead of 96% of the people. 
#In our result we got 83.24 percentile, it means our num 98 is ahead of 83.24 percent of the numbers.
```

```python
#median
a=np.random.randint(1,100,20)
print(a)

print(np.median(a))
```

```python
#Histogram

a=np.random.randint(1,100,20)
print(a)

np.histogram(a,bins=[0,10,20,30,40])  #I kind of created the buckets, now np.histogran will check which number 
#belong to which bucket and how many times.

```

```python
#corrcoef

salary=np.array([40000,20000,75000,14000,1000000])
experience=np.array([1,2,3,4,5])
np.corrcoef(salary,experience)

#+1 means strong positive correlation. -1 means strong negative correlation.
```

```python
#isin

#If we have two arrays, the isin function will tell whether the a number in one array present in another
#array or not.

a=np.random.randint(1,40,20)
print(a)
print("\n")
b=np.array([10,15,20,25,30,35,40,45,50])
print(b)
print("\n")
print(np.isin(a,b))
a[np.isin(a,b)]
```

```python
#flip -> Flip function flip the array in opposite direction
b=np.array([10,15,20,25,30,35,40,45,50])
print(b)
print("\n")
print(np.flip(b))
```

```python
b=np.array([[12, 52, 42, 6],
        [12, 18, 52, 55],
        [61, 93, 83, 93],
        [38, 61, 44, 85],
        [ 87, 87, 31, 2],
        [44, 71, 2, 7]])

print(np.flip(b))    #Flip both column wise and row wise
print("\n")
print(np.flip(b,axis=0))
print("\n")
print(np.flip(b,axis=1))
```

```python
#put function helps in chnaging the values in the array

b=np.array([10,15,20,25,30,35,40,45,50])
np.put(b,[1,4],[-15,-30])   #We have to provide the index value of the array we want to replace and the new values
print(b)

#Put is a permanent operation
```

```python
#Delete -> If we want to delete a specific item from the array

print(b)

np.delete(b,1)   #We just have to provide the index value 
```

#### Set Functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
#np.union1d

a=np.array([1,7,3,10,14,20])
b=np.array([12,11,8,1,10,7])

np.union1d(a,b)
```

```python
#np.intersect1d

np.intersect1d(a,b)
```

```python
#np.setdiff1d

np.setdiff1d(a,b)  #What is a has that b does not have
```

```python
#np.setxor1d

np.setxor1d(a,b)   #Shows all the elements that are present in both a and b, but not common elements
```

```python
#np.in1d

np.in1d(a,b)   #It is a kind of membership operator. It will tell what items of "b" "a" is having.
```

```python
print(np.in1d(a,1))   #It is showing whether 1 is present in a or not. 

a[np.in1d(a,1)]    #It will now directly showing the element
```
