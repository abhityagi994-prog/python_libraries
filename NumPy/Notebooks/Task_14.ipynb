```python
#Q-4: Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
#arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

arr=np.array([[1,2,np.nan],
              [4,2,6],
              [np.nan,np.nan,5]])

val=[arr[~np.isnan(arr)]]    #First we filter the values which are not nan
print(val)

values,count=np.unique(val,return_counts=True)    #Then we check the unique values and there count. return_count=True, will return the no. of values
common=values[np.argmax(count)]                   #occured. And then we find the maximum count no. corrosponding index value which is our common no.

arr[np.isnan(arr)]=2
print(arr)
```

```python
#Q-5: Write a NumPy program to find the missing data in a given array. Return a boolean matrix.
#also try to fill those missing values with 0. For that, you can use np.nan_to_num(a)
import numpy as np


arr=np.array([[3, 2, np.nan, 1],
          [10, 12, 10, 9],
          [5, np.nan, 1, np.nan]])

def is_nan(arr):
    nan_matrix=np.isnan(arr)

    arr[nan_matrix]=0
    return nan_matrix, arr
    

print(is_nan(arr))
```

```python
#Q-9: Write a program implement Boradcasting Rule to check if two array can be added or not.
#Given tuples of shapes.

#shape of a- (3,2,2)
#shape of b- (2,2)

#check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.


def Broadcasting(arr1,arr2):
    if len(arr1)<len(arr2):
        arr1=(1,)*(len(arr2)-len(arr1))+arr1
    elif len(arr2)<len(arr1):
        arr2=(1,)*(len(arr1)-len(arr2))+arr2

    for d1,d2 in zip(arr1,arr2):
        if d1!=d2 and d1!=1 and d2!=1:
            return False
    return True


def check_broadcast(a,b):
    return Broadcasting(np.shape(a),np.shape(b))

a=np.arange(1,13).reshape(3,2,2)
b=np.arange(1,5).reshape(2,2)

print(check_broadcast(a,b))
```
