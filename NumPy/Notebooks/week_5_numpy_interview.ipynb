```python
#meshgrids - It is basically used to plot the 3-D graphs. It prepares the data for 3D coordinates. 
```

```python
a=np.linspace(start=0,stop=9,num=10)
a.shape
```

```python
b=np.linspace(start=0,stop=9,num=10)
b.shape
```

```python
xx,yy=np.meshgrid(a,b)   ##It will now create a grid for both a and b numpy. 
print(xx)
print(yy)
```

```python
py.scatter(xx,yy)
```

```python
#What is the use of Grid we created before?

#We are preparing the data and can use the above grid and pass it to different functions which then can create a new coordinates
```

```python
print(xx.shape)
print(yy.shape)
```

```python
def func(a,b):
    return a**2+b**2

zz=func(xx,yy)
print(zz)
print(zz.shape)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=xx,y=yy,z=zz,mode='markers'))
fig.add_trace(go.Surface(x=xx, y=yy, z=zz))
fig.show()
```

```python
def func_plane(a,b):
    return 3*a+2*b

zz1=func_plane(xx,yy)
zz1
```

```python
#Seed - Seed kind of fixed the value of random. For example - If we run random.random or random.randint again and again, it will generate new set of 
#numbers every time. But if we use the seed, then everytime we are running rondom or randint, we will get the same set of numbers.
```

```python
np.random.seed(1)   #We can put any number in the seed.
np.random.random(size=(3,3))
```

```python
#Shuffle 

a=np.array([1,2,3,4,5])
print(a)
np.random.shuffle(a)   #Shuffle is a permanent operation
print(a) 
```

```python
a 
```

```python
#Choice

np.random.choice(a,3)   #It randomly take the items from the array. Over here the value is 3 so, it is taking out 3 numbers.
```

```python
#Also, the choice works like, it take the random item from the array and then again looks for a random item. So, there can be a possibility 
#that it might show the same repeated items. As, it extracts the element one by one.
np.random.choice(a,3)
```

```python
#We can overwrite the above behaviour by 
np.random.choice(a,3,replace=False)
```

```python
#Homogeneous or Heterogeneous arrays

a=np.array([1,'John',3.76,True])
print(type(a[0]))
print(type(a[1]))
print(type(a[-1]))
a
#As we can see the array got created but it is not the heterogeneous array. It is still a homogeneous array, and U32 means it can store 32 characters
```

```python
#But there is a way of creating heterogeneous arrays. Structured Arrays

data_type=np.dtype(
    [
        ('name','U20'),
        ('iq',np.int16),
        ('cgpa',np.float16),
        ('placed','U20')
    ]
)

data_type
```

```python
std=np.array(
    [
        ('John',100,91.3,'Yes'),
        ('Shane',112,93.6,'Yes'),
        ('Yon',68,75.4,'No')
    ],dtype=data_type
)
std

#In this way we can also create a heterogeneous array
```
