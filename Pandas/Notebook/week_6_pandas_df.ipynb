## Dataframe

- In pandas the dataframe is a 2D structure, made up of series. 
- We could say it is like a table and inside the table the individual columns are
the series.

```python
import pandas as pd
import numpy as np
```

### Creating a dataframe using Lists

```python
L=[
    ["John",1998,"100K"],
    ["Dio",2001,"650k"],
    ["Jake",2003,"120K"]
]

pd.DataFrame(L,columns=['name','DOB','salary'])
```

### Creating a dataframe using Dictionary

```python
d={
    "name":["John","Harry","Leo"],
    "DOB":[1998,2001,2003],
    "salary":["100K","650K","120K"]
}

pd.DataFrame(d)
```

```python
ipl[ipl["TossWinner"]==ipl["WinningTeam"]].shape[0]    #It means there are 489 matches where the Toss Winner Team are the Winning Team
```

```python
ipl[ipl["TossWinner"]==ipl["WinningTeam"]].shape[0]/ipl.shape[0]   #We are dividing it with total number of matches
```

```python
(ipl[ipl["TossWinner"]==ipl["WinningTeam"]].shape[0]/ipl.shape[0])*100
```

```python
#Action Movies with rating higher than 7.5
a=movies['genres'].apply(lambda x: x.split('|'))

idx=a[a.apply(lambda x: True if "Action" in x else False)].index
```

```python
df=movies.loc[idx]
df[df['imdb_rating']>7.5].shape[0]
```
