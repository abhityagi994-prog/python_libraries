```python
import pandas as pd
import numpy as np
```

```python
courses=pd.read_csv('courses.csv')
students=pd.read_csv('students.csv')
nov=pd.read_csv('reg-month1.csv')
dec=pd.read_csv('reg-month2.csv')

matches=pd.read_csv('matches.csv')
ipl=pd.read_csv('deliveries.csv')
```

```python
#pd.concat  - It is used to vertically stack the different data. 

print(pd.concat((nov,dec)).shape)
pd.concat((nov,dec),ignore_index=True)   #Ignore Index = True means it will ignore the original 
#index values of different dataframes
```

```python
#As it is retaining the original index values of Dataframe, we can provide our custom index

idx=np.arange(0,53)
pd.concat((nov,dec)).set_index(idx)
```

```python
#Multiindex dataframe - A MultiIndex DataFrame is a DataFrame where rows (or columns) have more than one label level instead of just one.
pd.concat((nov,dec),keys=['nov','dec'])
```

```python
months=pd.concat((nov,dec),keys=['November','December'])
months.loc['December',5]
```

```python
pd.concat((nov,dec),axis=1)
```

```python
#Merge or Join 

#Inner Join
std=pd.concat((nov,dec),ignore_index=True)
students.merge(std,how='inner',on='student_id')
```

```python
#Left Join
courses.merge(std,how='left',on='course_id')
```

```python
#Outer Join
students.merge(std,how='outer',on='student_id')
```

```python
#Find the total revenue generated by the company
std=pd.concat((nov,dec),ignore_index=True)
std.merge(courses,how='inner',on='course_id')['price'].sum()
```

```python
#Find month by month revenue

#Method-1
std_1=pd.concat((nov,dec),keys=('November','December'))
std_1.loc['November']['course_id'].sum()
```

```python
#Method-2

std_1=pd.concat((nov,dec),keys=('November','December')).reset_index()
std_1.merge(courses,how='inner',on='course_id').groupby('level_0')['price'].sum()
```

```python
#Print the registration table
#-> Columns to include student name, course, price

std=pd.concat((nov,dec),ignore_index=True)
std_1=std.merge(courses,how='inner',on='course_id')
std_1.merge(students,how='inner',on='student_id')[['name','course_name','price']]
```

```python
#Plot a bar chart of revenue and course
rev=pd.concat((nov,dec),ignore_index=True).merge(courses,how='inner',on='course_id')
revenue_generated = rev.groupby('course_name')['price'].sum().reset_index()

import matplotlib
from matplotlib import pyplot as plt

revenue_generated

plt.barh(revenue_generated['course_name'],revenue_generated['price'])
```

```python
#Find the students who enrolled in both the months

stds=nov.merge(dec,how='inner',on='student_id').merge(students,how='inner',on='student_id')
pd.unique(stds['name'])
```

```python
#Find the courses that got no enrollment

std=pd.concat((nov,dec),ignore_index=True)
na_courses=courses.merge(std,how='left',on='course_id')
na_courses[na_courses['student_id'].isna()][['course_id','course_name']]
```

```python
#Print the student name and their partner name

students.merge(students,how='inner',left_on='partner',right_on='student_id')   
#Always remember over here we are performing self join. Although we wrote the inner join, but it is a self join. 
#Also, left_on belong to the outer dataframe and right_on always belong to the inner dataframe.
```

```python
#Find top 3 students who did most number of enrollments

pd.concat((nov,dec)).merge(students,how='inner',on='student_id').groupby(['student_id','name']).size().sort_values(ascending=False).head(3)
```

```python
#Find top 3 students who spend most amount of money on courses

std_=pd.concat((nov,dec)).merge(students,how='inner',on='student_id')
std_.merge(courses,how='inner',on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
#Alternative Syntax for merge

pd.merge(nov,courses,how='inner',on='course_id')    #This is an alternative syntax for mergeing. The first dataframe is the left one
#And the second table is the right one.
```
