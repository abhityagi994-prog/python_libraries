```python
students={
    'name':["Nina","Tokyo",np.nan,"Shin",np.nan],
    'Age':[20,19,np.nan,np.nan,23],
}

df1=pd.DataFrame(students)
df1
```

```python
#Dropna

#On dataframe, it will remove the entire row that contains even a single nan value. 

df1.dropna()
#As we can see it removed the entire row that contain the single nan value
```

```python
#The above thing happened because of the default behaviour of dropna

df1.dropna(how="all")    #"how" is a hidden parameter. Usually how="any", now we set it to all. Which means only remove those rows which have 
#all the values NaN
```

```python
df1.dropna(how="any")
```

```python
#If I want to remove the rows based on a particular column

df1.dropna(subset='name')
```

```python
#If I want to remove the rows based on multiple column

df1.dropna(subset=['name','Age'])
```

```python
#Fillna for series

df1['name'].fillna('unknown')
```

```python
#Fillna for dataframe

df1.fillna(0)  #It fills the NaN value for the entire dataframe
```

```python
#forward fill and backword fill

df1['name'].ffill()   #It fills the NaN values with the forward fill
```

```python
df1['name'].bfill()
```

```python
#Apply function in Dataframe



points_df = pd.DataFrame(
    {
        '1st point': [(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point': [(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
#We want to find out the euclidean distance

def euclidean(row):
    pt_a=row['1st point']
    pt_b=row['2nd point']
    return ((pt_b[0]-pt_a[0])**2+(pt_b[1]-pt_a[1])**2)**0.5

points_df.apply(euclidean,axis=1)    #If we don't put the value axis=1, then it will go with it's default nature i.e. columns

#Also by simply passing the function in apply() pandas will call the function itself. And passes the rows one by one as an argument.
```
