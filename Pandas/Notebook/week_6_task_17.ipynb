Q-6 Write a code which can display the bar chart of top 5 teams who have played maximum number of matches in the IPL.

```python
s1=ipl["Team1"].value_counts(ascending=False)
s1
```

```python
s2=ipl["Team2"].value_counts(ascending=False)
s2
```

Q-8: Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.

team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')

```python
def dashboard(team1:str,team2:str):
    if not isinstance(team1,str) or not isinstance(team2,str):
        raise TypeError("Input should be in string")
    try:
        mask=ipl[(ipl['Team1']==team1) & (ipl['Team2']==team2) | (ipl['Team1']==team2) & (ipl['Team2']==team1)][['Team1','Team2','WinningTeam','Player_of_Match']]
        print(mask['WinningTeam'].value_counts())
        print(mask['Player_of_Match'].value_counts().head(1))
        print("Csk Win rate",(mask['WinningTeam'].value_counts().iloc[0]/mask['WinningTeam'].value_counts().sum())*100)
        print("KKr Win rate",(mask['WinningTeam'].value_counts().iloc[1]/mask['WinningTeam'].value_counts().sum())*100)
    except Exception as e:
        print(e)

dashboard('Kolkata Knight Riders','Chennai Super Kings')
```
