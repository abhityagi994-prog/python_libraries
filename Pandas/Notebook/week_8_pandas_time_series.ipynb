```python
import pandas as pd
import numpy as np
```

#### Timestamp Object 
- Timestamp refers to the particular moment or instance of time. For example - The time at the current moment can be called as a time stamp

```python
#Creating a Timestamp Object 
pd.Timestamp('2026/1/27')    #Always try to provide the input in the format Year, Month, Day
```

```python
#Various ways of providing the timestamp
print(pd.Timestamp('2026, 1, 27'))   #It is important to give the space while using the , for timestamp
pd.Timestamp('2026-1-27')
```

```python
#Using the text 
pd.Timestamp('27th January 2026')
```

```python
#Providing time also
pd.Timestamp('27th January 2026 6:54PM')    #Also, while defining the time we have to put the value in 1-12 format not 
#1-24 format or else, it will throw an error.
```

```python
#Using the date time
import datetime 
```

```python
dtime=datetime.datetime(2026 ,1 ,27 ,19 ,1)
dtime   #This is a datetime object. The datetime object belongs to the python
```

##### Why there was a need of creating a new Timestamp datatype in pandas when we have datetime object in python?

- Syntax wise datetime is very convenient
- But performance take a hit when we have a very large dataset. As we know list are slower than the nparray
- The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
- The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.

```python
dt_1=np.array('2026-01-27',dtype=np.datetime64)   
#Important -> Always remember while creatign a datetime object using the numpy the format should be '0000-00-00'
dt_1
```

#### Datetime Index
- When we want to store only one date, we use timestamp
- When we want to store multiple dates we use Datetime index

```python
pd.DatetimeIndex(['2026/01/25','2026/01/26','2026/01/27'])
```

```python
#Using the python datetime object to create the Datetimeindex

a=pd.DatetimeIndex([datetime.datetime(2026,1,27),datetime.datetime(2026,2,4)])
a
```

```python
#Creating a datetime index for a series
pd.Series([1,2],index=a)
```

```python
#Date Range function
pd.date_range(start='2026, 2, 26',end='2026/3/10',freq='D')   #Frequency D means day
```

```python
#If we want alternative days
pd.date_range(start='2026, 2, 26',end='2026/3/10',freq='2D') 
```

```python
#If we want the time period for months
pd.date_range(start='2026, 2, 26',end='2026/6/10',freq='ME') 
```

```python
#If we want only business days not weekends
pd.date_range(start='2026, 2, 26',end='2026/3/10',freq='B')
```

```python
#If we want one day a week
pd.date_range(start='2026, 2, 26',end='2026/3/20',freq='W') 
#By default it is showing sunday
```

```python
#If we want Thursday instead of Sunday
pd.date_range(start='2026, 2, 26',end='2026/3/10',freq='W-THU') 
pd.date_range(start='2026, 2, 26',end='2026/3/10',freq='W-FRI') 
#Now, it is showing the thursday dates
```

```python
#If we want year end 
pd.date_range(start='2026, 2, 26',end='2032/3/10',freq='YE') 
```

```python
#Using the Periods 
pd.date_range(start='2026, 2, 26',periods=15,freq='D') 
```

```python
#to_datetime function
s=pd.Series(['2026/1/24','2026/1/25','2026/1/26'])
s
```

```python
s=pd.to_datetime(s)
s
```

```python
s.dt.year
```

```python
#Handling the errors
b=pd.Series(['2026/1/24','2026/100/25','2026/1/26'])
b
```

```python
pd.to_datetime(b,errors='coerce')  #It will ignore the value that cannot be converted into the Datetime.
#NaT -> Not a Time
```

```python
pd.to_datetime(b,errors='coerce').dt.month_name()
```
