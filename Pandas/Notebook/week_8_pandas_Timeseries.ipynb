#### Timedelta Object 
- It is also an object of pandas just like timestamp, series, Dataframe
- Represents the duration (difference) between two dates or times

```python
#Calculating the time difference
t1=pd.Timestamp('26th Jan 2026 8:25:10')
t2=pd.Timestamp('27th Jan 2026 10:55:50')
t2-t1
#As we can see the result shows the timedelta
```

```python
#Creating a Timedelta Object
pd.Timedelta(2,'d')
#OR
d=pd.Timedelta(days=2,hours=10,minutes=30)
d
```

```python
#Arthimetic Operations
pd.Timestamp('2026-01-26')+d  #As we can see we can add the timedelta with the time stamp
```

```python
#On Multiple Timestamp

pd.date_range(start='26th Jan 2026',end='2nd Feb 2026')+d
```

Examples

. Financial Data (Company stocks)
. Natural Data (Rainfall measurement)
· Event Data (Covid)
. Medical Data (Heart rate monitoring)

Types of Operations done on Time Series

· Time Series Analysis
· Time Series Forecasting

### Resampling
Resampling involves changing the frequency of your time series observations.

Two types of resampling are:

- Upsampling: Where you increase the frequency of the samples, such as from minutes to seconds.
- Downsampling: Wbere vou decrease the frequency of the samples, such as from davs to months.

```python
google['Close'].plot()
#The data is on the basis of days frequency
```

```python
#Downsampling 
#Converting from days to week or months

google['Close'].resample('ME').mean()
#For resampling we have to provide whether we want to resample on the basis of months, years, weeks etc.
#Also, the resampling works like -> if we put the resample as month, then we have to provide the aggregation function as well, so that it can 
#combine all the values for a particular month, just like groupby
```

```python
google['Close'].resample('ME').mean().plot()
#The below graph now has less noice
```

```python
#Upsampling 
google.resample('h') #->It means by  hours
#But as we don't have the data for every hour. We will use the interpolate function and specify the function according to which 
#we want to fill the missing values
google.resample('4h').interpolate(method='linear')
```

### Rolling Window(Smoothing)

Time series data in original format can be quite volatile, especially on smaller aggregation levels. The concept of rolling, or moving averages is
a useful technique for smoothing time series data.

- For example let's say we have a DF with index as a date(per day) and close price as a column. Now, we decide the time period like 3 days or 5 days.
and according to that time period we take (Let's say 3) then mean (any agg function) of first 3 rows and then next 3 rows and so on...

```python
#Rolling window

google['Close'].rolling(4).mean()
#As we can see that we have taken the window period of 4 day. So, for the first 3 it is showing NaN, and then afterwards for every 4 days it is 
#showing the results.
#The reason it showed 3 NaN instead of 4 because pandas as soon as the condition meet for the 4th row, it calculated the mean.
#Upto the row 3 the condition was not satisfying, for the 4th row condition satisfied and it computer the mean of first 4 rows and assign
#it to the 4th row
```

```python
google['Close'].plot()
google['Close'].rolling(30).mean().plot()
```

```python
google['Close'].plot()
google['Close'].rolling(30).mean().plot()
google['Close'].ewm(30).mean().plot()
```
