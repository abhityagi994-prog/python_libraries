Q-8: Player with most appearance in final match.

```python
import ast

m1=ipl[ipl['MatchNumber']=='Final']
players=m1['Team1Players'].apply(ast.literal_eval)+m1['Team2Players'].apply(ast.literal_eval)
d={}
for player in players:
    for men in player:
        if men in d:
            d[men]+=1
        else:
            d[men]=1

df=pd.DataFrame(d.items(),columns=["Player","Frequency"])
df.sort_values('Frequency',ascending=False)
```

Q-9: IPL Point Table
Make a function point_table which take season as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits; For loosing - 0 Point For not result both team gets 1 points.

Make dataframe which will have TeamName MatchesPlayed MatchesWon NoResult Points make TeamName as index.

```python
def points_table(season,matches=ipl):
    df=ipl[ipl['Season']==str(season)]
    teams=pd.unique(pd.concat((df['Team1'],df['Team2'])))
    
    rows=[]
    
    for team in teams:
        main_team=df[(df['Team1']==team) | (df['Team2']==team)]
        play=main_team.shape[0]
        won=df[(df['WinningTeam']==team) &((df['Team1']==team) | (df['Team2']==team))].shape[0]
        no_result=(main_team['WinningTeam'].isna()).sum()

        points=won*2+no_result*1

        rows.append([team,play,won,no_result,points])

    new_df=pd.DataFrame(rows,columns=["Team Name","Matches Played","MatchesWon", "NoResult", "Points"]).set_index('Team Name')
    return new_df.sort_values('Points',ascending=False)

print(points_table(2015))
```
