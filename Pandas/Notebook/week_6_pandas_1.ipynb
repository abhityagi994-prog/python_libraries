```python
#Copy and View

#For example - 
a1=pd.Series([10,20,30,40,50,60])
a1
```

```python
new=a1.head(3)
new
```

```python
#Now, let's say we want to change something in new
new.iloc[2]=300
new
```

```python
#Now , if we print() the a1
a1
```

```python
#As we call the a1. We can say that change in new, did the changes in a1.
#This is becaus  new is still pointing to the a1

new=a1.head(3).copy()
new.iloc[2]=600
```

```python
a1
```

```python
new

#As we can see 600 still not updated.
```

```python
#Also, Important thing to remember is -> head and tail, these funtions provides the view of the data not the copy.
#Hence, when we did the changes in new without copy, the changes also reflect in the a1 as well.
```
