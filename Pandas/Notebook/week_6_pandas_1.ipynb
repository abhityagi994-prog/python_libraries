```python
#drop_duplicate will drop the duplicate values

a=pd.Series([1,1,2,2,2,3,3,4,4])
a.drop_duplicates()    #It always delete the first occurance
```

```python
a.drop_duplicates(keep='last')   #It is now keeping the last values
```

```python
#duplicated -> Will tell which values are duplicated. 

a.duplicated().sum()
#Total 5 duplicated values
```

```python
a=pd.Series([4,3,np.nan,12,17,np.nan])
print(a.isnull)
```

```python
#Count always gives the values that are not null. It exclude the null values
#Whereas size include the null values

print(a.count())
print(a.size)
```

```python
#isnull

a.isnull().sum()   #We will know the total number of missing values
```

```python
#How to handle the missing values

#We have two ways to deal with the missing values
#1. dropna
#2. fillna
```

```python
a.dropna()
```

```python
a.fillna(1)
```

```python
#We can fill the na values with any value

a.fillna(a.mean())   
```
