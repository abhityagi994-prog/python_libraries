#### Very Important Concept of Positional and Label indexing

- We have two types of Indexing: Positional which Pandas assign and Label indexing or custom indexing which we sometimes provides
- For k_runs[-1] we are getting error while for movies[-1], and it worked because here pandas didn't got confused because [-1] cannot be a label in strings. It can only be a positional.
- For virat_kohli_runs[-1] did't work becuase pandas got confused whether it is a label_indexing or positional indexing. That's why it throws an error.
- Negative Indexing works in pandas for only positional, because this indexing is created by pandas internall.
- Even if we have a custom indexing, pandas also assign the positional indexing, and it ranges from 0 to n-1.
- Negative Indexing for label will not work until we have custom label -1 in the indexing.
- Hence, if we want to fetch the last record from virat_kohli_runs, we should use iloc. It means positional indexing.
- And loc for fetching some specific record based on custom indexing.

```python
#Positional Indexing
#Last value from k_runs 

k_runs.iloc[-1]
```

```python
#Label Indexing
#Specific actor name using custom indexing(movie_name)

movies.loc['Why Cheat India']
```

### Editing a Series

```python
d={
    "chemistry":89,
    "english":78,
    "maths":94,
    "physics":98
}

d1=pd.Series(d)
```

```python
#Overriding a value

d1.loc['english']=88
d1
```

```python
#Creating a new value in series if not present

d1['bio']=96
d1
```

```python
a1=np.arange(11,21)
a_series=pd.Series(a1)
a_series
```

```python
a_series[6]=10
a_series
```

```python
#Overriding using Slicing
a_series[2:7]=1
a_series
```

```python
#Editing using fancy indexing

a_series[[2,4,6]]=100
a_series
```
